```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
data=load_iris()
X=data.data
y=data.target
```

```python
X.shape
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#applying PCA

from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_scaled=pca.fit_transform(X)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=y)
```

Agglomarative clustering >> make dendogram

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(25,10))   # bigger size for clarity
plt.title("Dendrogram")
dend = sc.dendrogram(
    sc.linkage(pca_scaled, method="ward"),
    truncate_mode="lastp",   # show only last p clusters
    p=12,                    # number of clusters to display
    leaf_rotation=90,        # rotate labels
    leaf_font_size=12,       # font size for readability
    show_contracted=True     # show contracted clusters
)
plt.xlabel("Sample data")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage="ward")
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coeff=[]

for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,linkage="ward")
    agglo.fit(X_scaled)
    score=silhouette_score(X_scaled,agglo.labels_)
    silhouette_coeff.append(score)
```

```python
silhouette_coeff
```

```python
plt.plot(range(2,11),silhouette_coeff)
plt.xticks(range(2,11))
plt.xlabel("No of cluster")
plt.ylabel("silhouette_coefficient")
```

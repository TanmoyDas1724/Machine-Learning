Python pickle module is used for serialising and de-seralising a python object structure. Any object in python can be pickled so taht it can be saved on disk . What pickle does is that it "Serializes" the object first before writing it to file . Pickling is a way to convert a python Objetc(list , dict, etc) into a character stream . The idea is that this character stream contains all the infomation necessary to reconstruct the objet in another python script

PICKLE >> allows you store the python object on the disk 

SERIALIZATION >> Serialization is the process of converting a data structure or object into a format that can be easily stored, transmitted, or persisted. During serialization, the object's state is converted into a stream of bytes or other formats that can be written to disk, sent over a network, or stored in memory

DESERIALIZATION >> Deserialization is the process of reconstructing a serialized object back into its original form. During deserialization, the serialized data is read and converted back into a data structure or object that matches the original.

```python
import pickle
pickle.dump(model,open("model.pkl","wb")) # wb >> write byte

# python's object(model) , its attribute and methods is converted in binary mode
```

```python
X_test
```

```python
model=pickle.load(open("model.pkl",'rb'))
```

DIVIDED THE DATA IN X AND Y

for simple linear regression we need only two feature x,y

```python
x=data[["bmi"]]
y=data["target"]
```

lmplot is a Seaborn function used to visualize the relationship between two variables with a regression line.
It creates a scatter plot + best fit regression line.

```python
import seaborn as sns
sns.lmplot(x="bmi", y="target", data=data)
```

SPLIT THE DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 , random_state=1)
```

MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
```

INSTANTIATE THE EMPTY LR MODEL

```python
model = LinearRegression()
```

```python
model
```

Traning of Model

```python
model.fit(x_train, y_train)   ##### MUST IN x_train IN DATAFRAME FORMAT
```

```python
model.coef_  #m=slope there coef_
```

```python
model.intercept_ ## C
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

VISUALIZATION

```python
import seaborn as sns
sns.lmplot(x="bmi", y="target", data=data)
```

```python
plt.scatter(x_test,y_test , color='black', label='Actual data')

plt.plot(x_test,y_predict,color="blue", linewidth=3 , label= "Linear Regression Line")

plt.xlabel("BMI")
plt.ylabel("One year progression-Target")

plt.title("Linear Regression on diabetic data")

plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_predict)
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

For Binary classification >> We drop a class which is 

```python
df_final=df[df['Target']!=2]
```

```python
df_final
```

```python
df_final.Target.unique()
```

```python
X=df_final.iloc[:,:-1]
y=df_final.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=1)
```

```python
X_train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(data=X_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

```python
X_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
```

```python
classifier
```

```python
classifier.fit(X_train_scaled,y_train)
```

```python
y_pred=classifier.predict(X_test_scaled)
```

```python
y_pred
```

```python
classifier.predict_proba(X_test_scaled) ## corresponding to higher prebability class is predicted,cutoff 0.5
```

## Evaluation Matrix

```python
from sklearn.metrics import confusion_matrix , accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

1. First two rows>> 0 and 1 > precision , recall , f1 score and no of data point of respective class
2. macro average >> for each class take simple average >> precision of class 0 and class 1 will be macro average

3. similarly for recall , f1 score and support

4. weighted average is used in case of class imbalanced it gives importance to performance on dominant class

```python
from sklearn.metrics import roc_curve , auc
```

```python
classifier.predict_proba(X_test_scaled)[:,1] # probability for class 1
```

```python
y_pedict_proba=classifier.predict_proba(X_test_scaled)[:,1]
```

```python
# Reciever operating characteristics curve 
# Input : y_test and y_predict_proba . It will returen TPR, FPR with different cutoff of probabilities
# how your model is able to distincuised between two class
fpr , tpr, thresholds=roc_curve(y_test,y_pedict_proba)
```

```python
fpr
```

```python
tpr
```

```python
thresholds
```

```python
# calculate auc score
roc_auc=auc(fpr,tpr)
```

```python
roc_auc
```

```python
# plot roc_auc curve

plt.figure(figsize=(8, 6))  # Create a new figure with a specified size
plt.plot(fpr, tpr, color='darkorange', linewidth=2, label='ROC curve (area = %0.2f)' % roc_auc)  # Plot ROC curve,
plt.plot([0, 1], [0, 1], color='navy', linewidth=2, linestyle='--')  # Plot the diagonal line representing random guessing
plt.xlim([0.0, 1.0])  # Set x-axis limits
plt.ylim([0.0, 1.05])  # Set y-axis limits
plt.xlabel('False Positive Rate')  # Set x-axis label
plt.ylabel('True Positive Rate')  # Set y-axis label
plt.title('Receiver Operating Characteristic (ROC) Curve')  # Set plot title
plt.legend(loc="lower right")  # Add legend to the plot
plt.show()  # Show the plot
```

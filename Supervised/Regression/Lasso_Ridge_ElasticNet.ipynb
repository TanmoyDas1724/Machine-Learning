The "mpg" dataset, which stands for "miles per gallon". It contains information about various car models and their characteristics, such as cylinders, displacement, horsepower, weight, acceleration, model year, origin, and miles per gallon (mpg) fuel efficiency.

Here's a brief explanation of each column:

mpg: Miles per gallon, representing the fuel efficiency of the car.

cylinders: Number of cylinders in the engine.
displacement: Engine displacement, the measure of the cylinder volume swept by all of the pistons of a piston engine.

horsepower: The power of the engine, typically measured in horsepower (hp).

weight: Weight of the car, often measured in pounds.
acceleration: Acceleration of the car from 0 to 60 miles per hour (mph) in seconds.

model year: Year of manufacturing of the car model.

origin: Origin of the car, represented as a categorical 
variable (1: USA, 2: Europe, 3: Japan).

name: The name of the car model.

This dataset is commonly used for regression tasks, where the goal is to predict the fuel efficiency (mpg) of a car based on its other characteristics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=sns.load_dataset('mpg')
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop("name", axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df['horsepower']=df['horsepower'].fillna(df["horsepower"].median())
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df["origin"]
```

```python
df["origin"].unique()
```

```python
df["origin"].value_counts()
```

```python
df['origin']=df['origin'].map({
   "usa":1,
   "japan":2,
   "europe":3 
})
```

```python
df.info()
```

```python
X=df.drop("mpg",axis=1)
y=df['mpg']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression_model=LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(X_train,y_train)
```

```python
for i , col_name in enumerate(X_train.columns):
    print(i,col_name)
```

```python
for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {regression_model.coef_[i]}")
     
```

COEFFICIENTS ARE RELATIVE SMALLER It means if one independent variable changes slightly there will bw not much  difference in prediction

$ This is some thime called SMOOTHER model

```python
from sklearn.metrics import r2_score
```

```python
y_pred_LinearModel=regression_model.predict(X_test)
```

```python
y_pred_LinearModel
```

```python
r2_Linear=r2_score(y_test,y_pred_LinearModel)
```

```python
print(f"The R square value for linear regreesion is {r2_Linear}")
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge_regression_model=Ridge(alpha=0.1)
ridge_regression_model.fit(X_train,y_train)

for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
     

```

```python
y_pred_RidgeModel=ridge_regression_model.predict(X_test)
r2_Ridge=r2_score(y_test,y_pred_RidgeModel)
print(f"The R square value for Ridge regreesion is {r2_Ridge}")


```

We dont see much variation in coefficient of ridge regression as compared to linear regression

Lasso

```python
from sklearn.linear_model import Lasso
```

```python
Lasso_regression_model=Lasso(alpha=0.5)
Lasso_regression_model.fit(X_train,y_train)

for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {Lasso_regression_model.coef_[i]}")
```

3 features coefficients are 0 , Lasso helps in feature selection

```python
y_pred_LassoModel=Lasso_regression_model.predict(X_test)
r2_Lasso=r2_score(y_test,y_pred_LassoModel)
print(f"The R square value for Ridge regreesion is {r2_Lasso}")
```

Elastic net

```python
from sklearn.linear_model import ElasticNet
```

```python
ElasticNet_regression_model=ElasticNet(alpha=1, l1_ratio=.5)
ElasticNet_regression_model.fit(X_train,y_train)

for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {ElasticNet_regression_model.coef_[i]}")
```

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train,y_train)

y_pred=lassocv.predict(X_test)



score=r2_score(y_test,y_pred)
print("R2 Score", score)
```

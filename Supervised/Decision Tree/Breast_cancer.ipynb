```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset=load_breast_cancer()
```

```python
print(dataset.DESCR)
```

```python
feature=dataset.feature_names
```

```python
feature
```

```python
df=pd.DataFrame(dataset.data,columns=feature)
```

```python
df
```

```python
dataset.target
```

```python
df['Target']=dataset.target
```

```python
df
```

```python
df.tail()
```

```python
X=df.iloc[:,:-1]
y=df['Target']
```

```python
X
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test , y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
from sklearn.tree import DecisionTreeClassifier
breast_classifier=DecisionTreeClassifier()
breast_classifier.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,20))
tree.plot_tree(breast_classifier,filled=True)
```

```python
breast_classifier =DecisionTreeClassifier(max_depth=1)
breast_classifier.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(breast_classifier,filled=True)
```

```python
y_pred=breast_classifier.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
